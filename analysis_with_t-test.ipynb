```python
# import necessary packages
import pandas as pd
from matplotlib import pyplot
from scipy import stats
import statistics
```

```python
# read in csv
df=pd.read_csv('data/joined_data.csv')
```

```python
# display dataframe
df
```

```python
# filter for just heart failure
heart_df=df[(df['APR DRG Description']=='Heart Failure')]
```

```python
# filter to severity level 1
heart_df_lvl_1=heart_df[heart_df['APR Severity of Illness Code']==1]
heart_df_lvl_1
```

```python
# filter to year 2016
heart_df_lv1_1_2016=heart_df_lvl_1[heart_df_lvl_1['Year']==2016]
heart_df_lv1_1_2016
```

```python
# display kings county heart failure data
kings_series_heart=heart_df_lv1_1_2016[heart_df_lv1_1_2016['County']=='Kings']['Mean Charge']
print(kings_series_heart.count())
print('mean:',round(kings_series_heart.mean(),2))
print('stdev:',round(statistics.stdev(kings_series_heart),2))
kings_series_heart.hist(bins=10)
```

```python
# display new york county heart failure data
new_york_series_heart=heart_df_lv1_1_2016[heart_df_lv1_1_2016['County']=='New York']['Mean Charge']
print(new_york_series_heart.count())
print('mean:',round(new_york_series_heart.mean(),2))
print('stdev:',round(statistics.stdev(new_york_series_heart),2))
new_york_series_heart.hist(bins=10)
```

```python
# run t test
t_stat, p_value = stats.ttest_ind(new_york_series_heart,kings_series_heart)
print('t-statistic:',round(t_stat,2))
print('p-value:',round(p_value,2))
```

```python
# filter for births
birth_df_lvl1=df[(df['APR DRG Description']=='Vaginal Delivery') & (df['APR Severity of Illness Code']==1)]
```

```python
# filter for year
birth_df_lvl1_2016=birth_df_lvl1[(birth_df_lvl1['Year']==2016)]
```

```python
# multi year display
birth_df_lvl1.groupby('Year')['Mean Charge'].describe().round(2)
```

```python
# new york county
new_york_series=birth_df_lvl1_2016[birth_df_lvl1_2016['County']=='New York']['Mean Charge']
print(new_york_series.count())
print('mean:',new_york_series.mean())
print('stdev:',statistics.stdev(new_york_series))
new_york_series.hist(bins=10)
```

```python
# kings county
kings_series=birth_df_lvl1_2016[birth_df_lvl1_2016['County']=='Kings']['Mean Charge']
print(kings_series.count())
print(kings_series.mean())
print('stdev:',statistics.stdev(kings_series))
kings_series.hist(bins=10)
```

```python
# t-test explained at:
# https://towardsdatascience.com/inferential-statistics-series-t-test-using-numpy-2718f8f9bf2f
t_stat, p_value = stats.ttest_ind(new_york_series,kings_series)
print('t-statistic:',round(t_stat,2))
print('p-value:',round(p_value,2))
```
